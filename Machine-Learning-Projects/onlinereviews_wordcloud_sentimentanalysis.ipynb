```python
# -*- coding: utf-8 -*-

# Importing Scrapy Library & CrawlerProcess to run the script in Jupyter
import scrapy
from scrapy.crawler import CrawlerProcess

# Creating a new class to implement Spider
class AmazonReviewsSpider(scrapy.Spider):

    # Spider name
    name = 'amazon_reviews'

    # Domain names to scrape
    allowed_domains = ['amazon.com']

    # Base URL for the apple Watch reviews
    myBaseUrl = "https://www.amazon.com/Apple-Watch-GPS-40mm-Aluminum/product-reviews/B07XR5TRSZ/ref=cm_cr_arp_d_paging_btm_next_2?pageNumber="
    start_urls=[]

    # Creating list of urls to be scraped by appending page number a the end of base url
    for i in range(1,121):
        start_urls.append(myBaseUrl+str(i))

    # Defining a Scrapy parser based on the HTML inspection from the reviews page
    def parse(self, response):
            data = response.css('#cm_cr-review_list')

            # Collecting product star ratings
            star_rating = data.css('.review-rating')

            # Collecting user reviews
            comments = data.css('.review-text')
            count = 0

            # Combining the results
            for review in star_rating:
                yield{'stars': ''.join(review.xpath('.//text()').extract()),
                      'comment': ''.join(comments[count].xpath(".//text()").extract())
                     }
                count=count+1

process = CrawlerProcess(settings={
    'FEED_FORMAT': 'CSV',
    'FEED_URI': 'apl_watch_reviews.csv'
})


process.crawl(AmazonReviewsSpider)
process.start() # the script will block here until the crawling is finished
```

```python
# importing the modules needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
# reading the csv file with apple watch reviews
df = pd.read_csv('apl_watch_reviews.csv')
```

```python
# checking out the contents of the dataframe
df.head()
```

```python
# removing the line change operators from the comment text
df.comment = df.comment.str.replace('\n','')
df.comment = df.comment.str.strip()
```

```python
def visualise_word_map(df):
    words=" "
    STOPWORDS.update(['watch','say','don','even','buy'])
    sw = set(STOPWORDS)
    for msg in df["comment"]:
        msg = str(msg).lower()
        words = words+msg+" "
    wc = WordCloud(width=3000, height=2500, background_color='white', stopwords = sw).generate(words)
    fig_size = plt.rcParams["figure.figsize"]
    fig_size[0] = 14
    fig_size[1] = 7
    plt.imshow(wc)
    plt.axis("off")
    plt.show()
```

```python
visualise_word_map(df)
```

```python
# importing TextBlob library for sentiment analysis
from textblob import TextBlob
# Defining a new column that will calculate the overall sentiment of the comment. Initializing it with empty
df['sentiment'] = 'empty'
```

```python
# k,l & m measure positive, neutral & negative sentiment per sentence respectively
# Count itarates through each row of the dataframe
k,l,m,count = 0,0,0,0
# iterating through each review 
for comments in df.iterrows():
# converting all letters to lower for TextBlob data pre-processing
    comments = str(comments[1]).lower()
# Defining separators to split the comments into individual sentences for sentiment polarity measurement
    each = comments.split('.' or ',' or '(' or ')')
# iterating through each sentence of the comment and measuring positive, neutral & negative sentiments per sentence
    for sentence in each:
        analysis = TextBlob(sentence)
        if analysis.sentiment.polarity > 0: 
            k = k + 1
        elif analysis.sentiment.polarity == 0:
            l = l + 1
        elif analysis.sentiment.polarity < 0:
            m = m + 1
    if m>k:
        df.loc[count,'sentiment'] = 'negative'
    else:
        df.loc[count,'sentiment'] = 'positive'
# Reinitialising k,l,m to zero for each new comment
    k,l,m = 0,0,0
# Increasing count value to match index number
    count = count + 1
```
